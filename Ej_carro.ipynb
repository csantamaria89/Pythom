```python
class Carros:
    def __init__(self,param_marca,param_color):
        self.marca = param_marca
        self.color = param_color
        
    def imprimir(self,param_valor):
        print (f'El vehiculo es de marca {self.marca} , de color {self.color}, vale {param_valor}')
        
```

```python
vehiculo = Carros("Renault","azul")
vehiculo.imprimir("6 palos")
```
